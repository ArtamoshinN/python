```python
def nearest_value(x, a):
    index = (np.abs(x - a)).argmin()
    return x[int(np.ceil(index // len(x[0]))), index - int(np.ceil(index // len(x[0]))) * len(x[0])]
```

```python
x = np.arange(0,10).reshape((2, 5))
a = 4.6
print(nearest_value(x, a))
```

```python
def construct_matrix(x, y):
    x = np.reshape(x, (5, 1))
    y = np.reshape(y, (5, 1))
    return np.concatenate((x, y), axis=1)
```

```python
def nonzero_product(x):
    a = np.diag(x)
    a = np.delete(a, np.where(a == 0)[0])
    if np.prod(a) == 0:
        return None
    else:
        return np.prod(a)
```

```python
x = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
print(nonzero_product(x))
```

8 задание

```python
def max_element(x):
    a = np.where(x == 0)[0]
    if x[-1] == 0:
        a = np.delete(a, -1)
    return max(x[a + 1])
```

```python
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
print(x)
print(max_element(x))
```

```python
import matplotlib.pyplot as plt
def gen_uncertain_data():
    x = np.linspace(0, 30, 100)
    y = np.sin(x/6*np.pi) + np.random.normal(0, 0.02, size=x.shape)
    error = np.random.normal(0.1, 0.02, size=y.shape)
    return x, y, error
```

```python
x, y, error = gen_uncertain_data()

def newPlot(x, y, error):
    y_er1 = y - error
    y_er2 = y + error
    plt.figure(figsize=(15,10))
    plt.title("Задание 15")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.plot(x, y, "-b", label="y(x)", linewidth=2.0)
    plt.plot(x, y_er1, color="black", label="y - error")
    plt.plot(x, y_er2, color="red", label="y + error")
    plt.fill_between(x, y_er1, y_er2, color="green")
    plt.legend(loc="upper left")
    plt.grid(True)

newPlot(x, y, error)
plt.show()
```
